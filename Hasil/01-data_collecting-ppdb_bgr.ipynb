```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from bs4 import BeautifulSoup
```

# Zonasi

```python
def scrape_data(url):
    driver = webdriver.Chrome()
    driver.get(url)

    time.sleep(10)
    wait = WebDriverWait(driver, 10)
    wait.until(EC.invisibility_of_element_located((By.CLASS_NAME, 'splash-screen')))

    all_data = []

    while True:
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        table = soup.find('table')

        # Mengambil elemen 'th' pertama sebagai nama kolom
        header_row = table.find('tr')
        headers = [header.get_text(strip=True) for header in header_row.find_all('th')]
        all_data.append(headers)

        for row in table.find_all('tr'):
            row_data = [cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])]
            all_data.append(row_data)

        next_button = driver.find_element(By.XPATH, '//button[@aria-label="next page"]')
        is_enabled = next_button.is_enabled()
        if not is_enabled:
            break

        time.sleep(3)
        next_button.click()
        time.sleep(5)

    driver.quit()

    # Membuat DataFrame dari all_data dengan menggunakan headers sebagai nama kolom
    df = pd.DataFrame(all_data[1:], columns=all_data[0])

    # Menghapus baris pertama
    df = df.drop(df.index[0]).dropna().set_index(['No'])
    
    return df
```

```python
# Membuat list kosong untuk menyimpan dataframe
dataframes = []

# Melakukan iterasi dari smp1 hingga smp20
for i in range(1, 21):
    filename = f"datasets/smp{i}_zonasi_hasil.csv"
    df = pd.read_csv(filename)
    dataframes.append(df)

# Menggabungkan dataframe menjadi satu dataframe
smp = pd.concat(dataframes).reset_index().drop(['index', 'No'], axis=1)

# Menampilkan dataframe hasil penggabungan
smp.head()
```

```python
# Menghapus 'm' pada kolom 'Jarak (meter)' dan mengganti koma dengan titik dan titik dengan koma
smp['Jarak (meter)'] = smp['Sekolah Pilihan'].apply(lambda x: x.split('(')[1].split(')')[0].strip()).str.replace('m', '').str.replace('.', '?').str.replace(',', '.').str.replace('?', '').astype(float)

# Membuat kolom baru 'Pilihan Sekolah', 'Zona', dan 'Jarak (meter)'
smp['Pilihan Sekolah'] = smp['Sekolah Pilihan'].apply(lambda x: x.split('-')[0].strip())
smp['Zona'] = smp['Sekolah Pilihan'].apply(lambda x: x.split('-')[1].split('(')[0].strip())
smp = smp.drop(['Sekolah Pilihan'], axis=1)
smp.head(10)
```

```python
smp.to_csv('datasets/seluruh_smp_hasil.csv')
```

```python
summary = smp[['Pilihan Sekolah', 'Jarak (meter)', 'Zona']].groupby(['Pilihan Sekolah', 'Zona']).agg({
    'Pilihan Sekolah': 'count',
    'Jarak (meter)': 'max'
})
summary.columns = ['Jumlah Pendaftar', 'Jarak Maksimal (meter)']

summary
```

```python
summary.to_csv('datasets/summary_ppdb_hasil.csv')
```
