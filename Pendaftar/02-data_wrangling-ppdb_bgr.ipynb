```python
import pandas as pd
```

```python
# Membuat list kosong untuk menyimpan dataframe
dataframes = []

# Melakukan iterasi dari smp1 hingga smp20
for i in range(1, 21):
    filename = f"datasets/smp{i}_zonasi.csv"
    df = pd.read_csv(filename)
    df['Jarak (meter)'] = df['Sekolah Pilihan 1'].apply(lambda x: x.split('(')[1].split(')')[0].strip()).str.replace('m', '').str.replace('.', '?').str.replace(',', '.').str.replace('?', '').astype(float)

    # Membuat kolom baru 'Pilihan Sekolah', 'Zona', dan 'Jarak (meter)'
    df['Pilihan Sekolah'] = df['Sekolah Pilihan 1'].apply(lambda x: x.split('-')[0].strip())
    df['Zona'] = df['Sekolah Pilihan 1'].apply(lambda x: x.split('-')[1].split('(')[0].strip())
    df = df.drop(['Sekolah Pilihan 1'], axis=1).sort_values(by='Jarak (meter)')

    dataframes.append(df)

# Menggabungkan dataframe menjadi satu dataframe
smp = pd.concat(dataframes).reset_index().drop(['index', 'No', 'Sekolah Pilihan 2'], axis=1)

# Menampilkan dataframe hasil penggabungan
smp.head()
```

```python
smp.to_csv('datasets/seluruh_smp.csv')
```

```python
dayatampung = pd.read_csv('datasets/smp_dayatampung.csv', index_col=[0])
dayatampung
```

```python
kota = 0.9
luar_kota = 0.1

zona = {
    'Zona 1': 0.65 * kota,
    'Zona 2': 0.20 * kota,
    'Zona 3': 0.10 * kota,
    'Zona 4': 0.05 * kota,
    'Zona 5': 0.55 * luar_kota,
    'Zona 6': 0.30 * luar_kota,
    'Zona 7': 0.15 * luar_kota
}

daya_tampung = {
    'SMP NEGERI 1 BOGOR': 141,
    'SMP NEGERI 2': 158,
    'SMP NEGERI 3 BOGOR': 141,
    'SMP NEGERI 4 BOGOR': 158,
    'SMP NEGERI 5': 158,
    'SMP NEGERI 6': 158,
    'SMP NEGERI 7 BOGOR': 158,
    'SMP NEGERI 8 BOGOR': 158,
    'SMP NEGERI 9 BOGOR': 158,
    'SMP NEGERI 10 BOGOR': 141,
    'SMP NEGERI 11 BOGOR': 158,
    'SMP NEGERI 12 BOGOR': 158,
    'SMP NEGERI 13 BOGOR': 123,
    'SMP NEGERI 14 BOGOR': 158,
    'SMP NEGERI 15 BOGOR': 158,
    'SMP NEGERI 16': 158,
    'SMP NEGERI 17 BOGOR': 141,
    'SMP NEGERI 18 KOTA BOGOR': 158,
    'SMP NEGERI 19 BOGOR': 158,
    'SMP NEGERI 20': 141
}
```

```python
def filter_and_save_smp_data(smp_name):
    smp_zona = pd.DataFrame()

    for zona_key, zona_value in zona.items():
        smp_zona_temp = smp.loc[(smp['Pilihan Sekolah'] == smp_name) & (smp['Zona'] == zona_key)].sort_values(['Zona', 'Jarak (meter)'])
        smp_zona_temp = smp_zona_temp.head(round(daya_tampung[smp_name] * zona_value))
        smp_zona = pd.concat([smp_zona, smp_zona_temp])

    csv_filename = f'datasets/zonasi_terfilter/{smp_name}.csv'
    smp_zona.to_csv(csv_filename)

    # Mengambil informasi jumlah baris dan kolom dari DataFrame
    rows, columns = smp_zona.shape

    print(f"{csv_filename} dibuat. Bentuk dataframe: ({rows}, {columns})")
```

```python
# Memanggil fungsi untuk setiap SMP
for smp_name in daya_tampung.keys():
    filter_and_save_smp_data(smp_name)
```

```python
smp_data = {}

def filter_and_save_smp_data(smp_name):
    smp_zona = pd.DataFrame()

    for zona_key, zona_value in zona.items():
        smp_zona_temp = smp.loc[(smp['Pilihan Sekolah'] == smp_name) & (smp['Zona'] == zona_key)].sort_values(['Zona', 'Jarak (meter)'])
        smp_zona_temp = smp_zona_temp.head(round(daya_tampung[smp_name] * zona_value))
        smp_zona = pd.concat([smp_zona, smp_zona_temp])

    smp_data[smp_name] = smp_zona
```

```python
# Memanggil fungsi untuk setiap SMP
for smp_name in daya_tampung.keys():
    filter_and_save_smp_data(smp_name)

# Menggabungkan semua DataFramenya menjadi satu DataFrame
combined_df = pd.concat(smp_data.values()).reset_index().drop(columns=['index'])
```
