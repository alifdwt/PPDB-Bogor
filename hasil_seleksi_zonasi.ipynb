```python
import pandas as pd
```

```python
# Baca data dari file CSV ke dalam dataframe
zonasi_pendaftar = pd.read_csv('Pendaftar/datasets/seluruh_smp.csv', index_col=[0])
zonasi_hasil = pd.read_csv('Hasil/datasets/seluruh_smp_hasil.csv', index_col=[0])

# Lakukan left join pada kolom "Nama Siswa" dan "Pilihan Sekolah"
merged_df = pd.merge(zonasi_pendaftar, zonasi_hasil, on=["Nama Siswa", "Pilihan Sekolah"], how="left")

# Buat kolom baru "Hasil" berdasarkan kondisi "Diterima" atau "Ditolak"
merged_df["Hasil"] = merged_df.apply(lambda row: "Diterima" if not pd.isna(row["Nomor Pendaftaran"]) else "Ditolak", axis=1)

# Hapus kolom "Nomor Pendaftaran" agar tidak ada duplikasi
merged_df.drop("Nomor Pendaftaran", axis=1, inplace=True)

# Tampilkan dataframe hasil
merged_df
```
